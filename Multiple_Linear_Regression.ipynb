Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

Data set

```python
data = pd.read_csv('/content/008 1.02.Multiple-linear-regression.csv')
```

```python
data.head()
```

```python
data.describe()
```

Explore Data

```python
y = data['GPA']
x = data [['SAT','Rand 1,2,3']]
```

```python
plt.scatter(data['SAT'],y, c = data['Rand 1,2,3'],cmap = 'jet')
plt.colorbar()
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

Regression

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
beeta = reg.coef_
beeta
```

```python
beeta0 = reg.intercept_
beeta0
```

```python
reg.score(x,y)
```

regression lines

```python
ybar = beeta0 + beeta[0]*data['SAT'] + beeta[1]*data['Rand 1,2,3']
ybar
```

```python
# rand = 1
y1 = beeta0 + beeta[0]*data['SAT'] + beeta[1]*1
y1
```

```python
plt.scatter(data['SAT'],y, c = data['Rand 1,2,3'],cmap = 'jet')
plt.colorbar()

plt.xlabel('SAT')
plt.ylabel('GPA')
fig = plt.plot(data['SAT'],y1,color = 'orange')
plt.show()
```

```python
# rand = 2
y2 = beeta0 + beeta[0]*data['SAT'] + beeta[1]*2
plt.scatter(data['SAT'],y, c = data['Rand 1,2,3'],cmap = 'jet')
plt.colorbar()

plt.xlabel('SAT')
plt.ylabel('GPA')
fig = plt.plot(data['SAT'],y2,color = 'red')
plt.show()
```

```python
# rand = 3
y3 = beeta0 + beeta[0]*data['SAT'] + beeta[1]*3
plt.scatter(data['SAT'],y, c = data['Rand 1,2,3'],cmap = 'jet')
plt.colorbar()

plt.xlabel('SAT')
plt.ylabel('GPA')
fig = plt.plot(data['SAT'],y3,color = 'green')
plt.show()
```

```python
y1 = beeta0 + beeta[0]*data['SAT'] + beeta[1]*2
plt.scatter(data['SAT'],y, c = data['Rand 1,2,3'],cmap = 'jet')
plt.colorbar()

plt.xlabel('SAT')
plt.ylabel('GPA')
fig = plt.plot(data['SAT'],y1,color = 'orange')
fig = plt.plot(data['SAT'],y2,color = 'red')
fig = plt.plot(data['SAT'],y3,color = 'green')
plt.show()
```
